```python
data = {
    'ins_lat':'37.56682', # 지정한 위도와 경도에서 가까운 순으로 나열
    'ins_lng':'126.97865',
    'p_sido_cd':'01', # 01=서울시, 08=경기 ... 16=제주
    'p_gugun_cd':'',  # 세부지역 (지정하지 않으면 시/도 전체)
    'in_biz_cd':'',
    'set_date':'',
    'iend':'1000',
}
    
url = 'https://www.starbucks.co.kr/store/getStore.do'
r = requests.post(url, data=data) #data는 딕셔너리 형태로 되어있음,

r.text[:1000] # 수신된 데이터의 앞부분만 확인
```

```python
jo=json.loads(r.text)
df=pd.json_normalize(jo,'list')
```

```python
len(df)
```

```python
df.columns #DataFrame 컬럼 인덱스 확인
```

```python
df = df[['s_name', 'lat', 'lot', 'sido_name', 'gugun_name', 'doro_address', 'tel']] #DataFrame 피처추출 
df.head(20)
```

```python
df.dtypes #데이터프레임의 데이터 타입 알려주는 함수
```

```python
df['lat']=df['lat'].astype(float)# pandas데이터를 형변환 하는법
df['lot']=df['lot'].astype(float)
```

```python
df.dtypes
```

```python
import folium
map_osm=folium.Map(location=(37.56629,126.979808)) #지도데이터 생성(lat,lot)
map_osm
```

```python
map_osm=folium.Map(location=(37.56629,126.979808),zoom_start=30) #지도확대
map_osm
```

```python
map_osm=folium.Map(location=(37.56629,126.979808),zoom_start=30,tiles='Stamen Toner') #타일 타입도 설정이 가능하다
map_osm
```

```python
시청_좌표=(37.56629,126.979808)
map_osm=folium.Map(location=시청_좌표,zoom_start=30,tiles='Stamen Toner') #타일 타입도 설정이 가능하다
```

```python
folium.Marker(시청_좌표,popup='시청').add_to(map_osm) #지도 위에 마킹해주는것
map_osm
```

```python
map_osm=folium.Map(location=시청_좌표,zoom_start=11)

#iterrow메소드는 인덱스와 행을 같이 던져주는 메소드
for ix,row in df.iterrows():
    location=(row['lat'],row['lot'])
    folium.Marker(location, popup=row['s_name']).add_to(map_osm)
map_osm
```
