```python
from nltk.corpus import names
import nltk
```

```python
nltk.download('names') #NLTK db다운로드
```

```python
labeled_names=([(name,'남자') for name in names.words('male.txt')]+\
               [(name,'여자') for name in names.words('female.txt')])
```

```python
labeled_names #(x data, ydata) , 입력데이터 정답데티어
```

```python
def gender_features(word):
    return {'last_letter':word[-1]}
```

```python
gender_features('Sopi')
```

```python
import random
random.shuffle(labeled_names)
labeled_names[0:10]
```

```python
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]
featuresets[0:10]
```

```python
train_set,test_set=featuresets[2000:],featuresets[0:2000]
```

```python
classifier=nltk.NaiveBayesClassifier.train(train_set) #학습결과 저장
```

```python
classifier.classify(gender_features('Sephipa'))
```

```python
classifier.classify(gender_features('Maria'))
```

```python
classifier.classify(gender_features('adam'))
```

```python
classifier.show_most_informative_features(5) #학습근거 
```

```python
nltk.classify.accuracy(classifier, test_set)
```
